```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pymongo
import json
```

```python
client = pymongo.MongoClient()
db = client.Twitter
tweets = db.raw_tstream
tweets.find_one()
```

```python
langs = {}
for x in tweets.find():
    try:
        lang = x['user']['lang']
    except KeyError:
        lang = 'unknown'
    if lang in langs:
        langs[lang]+=1
    else:
        langs[lang]=1
```
